### Challenge 1: The Normalizer

```python
np.random.seed(1)

data = np.random.randint(0, 100, (10, 4))  # 10 samples, 4 features
_min = data.min(axis=0)
_max = data.max(axis=0)

data_scaled = (data - _min) / (_max - _min)
data_scaled
```

```python
data_scaled.min(axis=0)
data_scaled.max(axis=0)
```

### Challenge 3: The Outlier Finder

```python
data = np.random.randn(100)
```

```python
mask = np.abs(data - data.mean()) > (2 * data.std())
mask
```

```python
mean_before = data.mean()
mean_before

data[mask] = 99.9
mask.sum()  # Count how many values were replaced

mean_after = data.mean()
mean_after
```

```python
data = np.array([5, 10, 30, 40, 50, 60, 1, -10])
mask = np.abs(data - data.mean()) > data.std()  # elem too far away from mean
data[mask] = 1000
mask.sum()  # these many you replaced
```

### Challenge 4: Image-Style Manipulations (Channel Swapping)

```python
data = np.random.randint(0, 256, (10, 10, 3))
```

```python
# "Mute" the Green channel by setting all its values to 0.
data[:, :, 1] = 0
```

```python
# Check if any pixel has a Red value greater than 200 and a Blue value less than 50.
mask = (data[:, :, 0] > 200) & (data[:, :, 2] < 50)
mask.sum()
```

```python
# average "brightness" of the image
data.mean()
```
