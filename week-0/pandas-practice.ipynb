```python
import pandas as pd
```

Pandas has Series and DataFrames
- Series: like an array
- DataFrame: like a 2d array (matrix, having rows and cols)

```python
data = {
    "name": ["rahul", "vidu", "vishnu"],
    "marks": [100, 98, 78],
    "age": [22, 21, 10],
}

df = pd.DataFrame(data, index=[90, 91, 93])
df
df.loc[90]  # exact index label
df.iloc[0]  # positional index (0, 1, 2, ...)
```

```python
df = pd.DataFrame(data)
```

```python
df.index
df.columns
df.values
```

```python
df['name']
type(df['name'])
```

```python
df[['name', 'marks']]
type(df[['name', 'marks']])
```

```python
df['name'][0]
df['name'][0:2]
```

```python
# df.loc[row_label, col_label] -> Scalar/Series/Df
# label: scalar, list, range
df.loc[0]
df.loc[0, 'name']
df.loc[0, ['name', 'age']]
df.loc[0:2, ['name', 'age']]
df.loc[0:2, 'name':'age']
df.loc[[0, 2], 'name':'age']
```

```python
# same as .loc but labels can only be numeric
df.iloc[0]
df.iloc[0, 0]  # df.iloc[0, 'name']
df.iloc[0, :2]
df.iloc[:, :2]
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)'

# Pass the 'User-Agent' header to mimic a browser
tables = pd.read_html(
    url,
    storage_options={'User-Agent': 'Mozilla/5.0'}
)

df_population = tables[0]
df_population
```

```python
mask = df_population['Population (1 July 2022)'] < 10_000
mask.sum()
df_population[mask]

mask2 = mask & (df_population['UN continental region[1]'] == 'Americas')
df_population[mask2]
```

```python
df = pd.DataFrame({
    "name": ['rahul', 'vidu', 'vishnu', 'paru', 'maa', 'papa', None],
    "age": [22, 21, None, 15, 42, 48, None],
    "height": [None, 150, 120, 165, 155, None, None]
})

df
df.info()
df.isnull().sum()
```

```python
df
df.dropna() # inplace=True
df.fillna(0)
```

```python
df['name'].fillna('Unknown')
```

```python
df.iloc[:, 1:].fillna(0)

df.iloc[:, 1:].fillna({
    "age": 20,
    "height": 150
})

df.iloc[:, 1:].fillna({
    "age": df['age'].mean(),
    "height": df['height'].mean(),
})
```

```python
df['power'] = 10
df
```

```python
df['power'] = df['age'] * df['height']
df
```

```python
df.columns
df.dtypes
df.select_dtypes(include=['number'])  # numeric cols
df.select_dtypes(include=['object'])  # categorical cols
```

```python
df = pd.DataFrame({
    "name": ['rahul', 'harikesh', 'paru', 'vishnu', 'vidu'],
    "place": ['bhadohi', 'bhadohi', 'hbh', 'hbh', 'hbh'],
    "study": ['bsc', 'bsc', 'school', 'school', 'ba'],
    "salary": [200, 400, 100, 40, 302]
})

df
```
