`.data` files have space (1 or many) separated columns

Because the separator is not fixed, `pd.read_csv()` will fail

We need to manually convert it to a proper CSV

```python
import re
```

```python
with open('data/housing.data') as file:
  data = file.read()
```

```python
header = ','.join(['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
                   'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV',])
```

```python
content = header + '\n'
for line in data.splitlines():
  content += re.sub(r'\s+', ',', line.strip()) + '\n'
```

```python
with open('data/housing.csv', 'w') as file:
  file.write(content)
```
