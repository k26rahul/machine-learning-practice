```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([
    ['2022-01-01', 'A', 10, 'rahul'],
    ['2022-01-01', 'B', 20, 'rahul'],
    ['2022-01-01', 'B', 50, 'rahul'],
    ['2022-01-02', 'A', 30, 'rahul'],
    ['2022-01-02', 'B', 40, 'rahul'],
    ['2022-01-02', 'C', 10, 'vidu'],
    ['2022-01-02', 'C', 30, 'vidu'],
    ['2022-01-02', 'C', 30, 'vidu'],
    ['2022-01-03', 'D', 20, 'vidu'],
], columns=['Date', 'Category', 'Qty', 'Customer'])
```

```python
table = df.pivot_table(values='Qty', index='Date', columns='Category', aggfunc='sum')
table
```

```python
type(table)
table.index
table.columns
```

```python
df.pivot_table(values='Qty', index='Date', columns='Category', aggfunc='count')
```

```python
df.pivot_table(values='Category', index='Qty', columns='Date', aggfunc='count', fill_value=0)
```

```python
table = df.pivot_table(values='Qty', index=['Customer', 'Category'],
                       columns='Date', aggfunc='sum', fill_value=0)
table
```
