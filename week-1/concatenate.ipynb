```python
import pandas as pd
```

```python
temp_data = pd.DataFrame({
    'Date': ['12-02-2023', '13-02-2023', '14-02-2023', '15-02-2023', '16-02-2023', '20-02-2023'],
    'TempMax': [24.3, 26.9, 23.4, 15.5, 16.1, 15]
}, index=range(17, 23))

rainfall_data = pd.DataFrame({
    'Date': ['12-02-2023', '13-02-2023', '14-02-2023', '15-02-2023', '16-02-2023', '25-02-2023'],
    'Rainfall': [0, 3.6, 3.6, 39.8, 2.8, 5]
}, index=range(85, 91))

temp_data
rainfall_data
```

```python
pd.concat([temp_data, rainfall_data])  # axis=0 by default, vertically concatenated
```

```python
pd.concat([temp_data, rainfall_data], ignore_index=True)
```

```python
pd.concat([temp_data, rainfall_data], axis=1)  # horizontally concatenated
```

# Join

```python
pd.concat([temp_data, rainfall_data], axis=0, join='inner')
```

```python
pd.concat([temp_data, rainfall_data], axis=0, join='outer')
```

# Keys

```python
df_keyed = pd.concat([temp_data, rainfall_data], axis=0, keys=('Delhi', 'Chennai'))
df_keyed
```
